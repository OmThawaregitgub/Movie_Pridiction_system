```python
import ast # Help to convert str to list
```

```python
def converter(text,keyword='name'): #Function that convert the str to list using ast.
    l=[]
    for i in ast.literal_eval(text): #apply loop to extract the str frome dataset and convert it into list
        l.append(i[keyword]) #Appending name from the list of dictionary.

    return l
```

```python
# Applying the convert function on the column.
merge_dataset['genres'] = merge_dataset['genres'].apply(converter)
```

```python
merge_dataset.sample(5) # Sample of the updated column.
```

```python
merge_dataset["keywords"][0]
```

```python
merge_dataset["keywords"] = merge_dataset["keywords"].apply(converter)
```

```python
merge_dataset["cast"][0]
```

```python
merge_dataset['cast'] = merge_dataset['cast'].apply(lambda x: converter(x,keyword = "character"))
```

```python
merge_dataset.sample(5)
```

```python
merge_dataset['crew'][0]
```

```python
def fetch_director(text):
    l=[]
    for i in ast.literal_eval(text):
        if(i['job'] == "Director"):
            l.append(i['name'])
            break
    return l
```

```python
merge_dataset['crew'] = merge_dataset['crew'].apply(fetch_director)
```

```python
merge_dataset.sample(5)
```

```python
#Converting the overview to list
merge_dataset["overview"] = merge_dataset["overview"].apply(lambda x:x.split()) 
```

```python
merge_dataset["overview"][0]
```

```python
# Function to remove extra space from the list.
def remove_space(word):
    l = []
    for i in word:
        l.append(i.replace(" ","")) #Replace space 
    return l
```

```python
merge_dataset['cast'] = merge_dataset['cast'].apply(remove_space)
merge_dataset['crew'] = merge_dataset['crew'].apply(remove_space)
merge_dataset['genres'] = merge_dataset['genres'].apply(remove_space)
merge_dataset['crew'] = merge_dataset['crew'].apply(remove_space)
merge_dataset['keywords'] = merge_dataset['keywords'].apply(remove_space)
```

```python
merge_dataset.sample(5)
```

```python
#Merging similar type conumn to one catagories.
merge_dataset['tags'] = merge_dataset['overview'] + merge_dataset['genres'] + merge_dataset['keywords']  + merge_dataset['cast'] + merge_dataset['crew']
merge_dataset[["movie_id",'title']] = credit[["movie_id",'title']]
```

```python
merge_dataset.head(2)
```

```python
new_df = merge_dataset[["movie_id",'title','tags']]
```

```python
new_df.dropna(inplace=True)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
# new_df['tags'] = new_df['tags'].apply(lambda x: x.replace(" ",""))
```

```python
new_df.head()
```

```python
new_df.isnull().sum()
```

```python
# Converting tages in lower case
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower()) 
```

```python
new_df['tags'][0]
```

```python
new_df.sample(4)
```

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
'''
Our column contain so many word that same meaning but different speling so we make function for
that word we identify that word and ake same spelling of the word. 
'''
def stems(text):
    l=[]
    for i in text.split():
        l.append(ps.stem(i))
    return " ".join(l)
```

```python
new_df['tags'] = new_df['tags'].apply(stems)
```

```python
new_df.head(5)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
def recommend(movie):
    index = new_df[new_df['title']==movie].index[0]
    distance = sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])
    for i in distance[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend("Kick-Ass")
```
